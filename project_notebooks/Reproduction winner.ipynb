```python
def estimateF0_autoCorr(x_win,fs,minF0 = 100,maxF0 = 1000):
    '''F0 detection on a single frame using autocorrelation
    Parameters
    ----------
    x_win : numpy.array
        Windowed signal frame
    fs,minF0,maxF0 : int
        Sampling rate, minimum and maximum F0 limits
        
    Returns
    -------
    ValAC : numpy.array
    f0 : numpy.array
        Estimated f0 and autocorrelation values for those f0
    ''' 
    f0 = np.array([])
    minT0 = int(fs/maxF0)
    maxT0 = int(fs/minF0)

    Ts = range(minT0,maxT0)
    ValAC = np.array([])
    for k in Ts:
        x_win_shifted=np.hstack((np.zeros(k),x_win[:-k]))
        autoCorr=np.dot(x_win,x_win_shifted)
        ValAC = np.append(ValAC, autoCorr)
        
    f0=np.divide(fs*np.ones(len(Ts)),Ts)
    return ValAC,f0
```

```python
def apply_mask(signal, mask):
    '''
        input:
            signal : numpy.array
            mask : numpy.array containing 0 for remove, 1 for keep
            
        output:
            numpy.array of the filtered signal
    '''
    return np.array(signal)*np.array(mask)
```

```python
def is_harmonic(f0,freq):
    ratio = freq/f0
    if ratio <1.01 : return False, -1
    if abs(ratio-round(ratio))<0.01 and round(ratio)!=1:
        return abs(ratio-round(ratio))<0.01 , round(ratio)
    else:
        return False, -1
```

```python
bool_mat = []
freqs_to_del_mat = []

for i,frame in enumerate(f_ac_arr):
    if i==0 or i==(len(f_ac_arr)-1): continue
    prev_frame = f_ac_arr[i-1]
    next_frame = f_ac_arr[i+1]
    local_bool = []
    freqs_to_del = []
    for freq in frame:
        for f0 in prev_frame:
            prev_bool,_ = is_harmonic(f0,freq)
            if prev_bool:
                freqs_to_del.append((f0,freq))
                break
        for f0 in next_frame:
            next_bool,_ = is_harmonic(f0,freq)
            if prev_bool:
                freqs_to_del.append((f0,freq))
                break
        local_bool.append(prev_bool or next_bool)
        if prev_bool or next_bool:
            print("freqs_to_del: ", freqs_to_del)
            print("frame: ", i)
            print("__________________________________________")
    freqs_to_del_mat.append(freqs_to_del)
    bool_mat.append(local_bool)
        
    #assert False
```
